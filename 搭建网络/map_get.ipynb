```python
import osmnx as ox
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
G = ox.graph_from_place(',China',which_result=2,simplify=False,
                        custom_filter=('["area"!~"yes"]["highway"~"ringway|primary"]'
                                      '["service"!~"parking|parking_aisle|driveway|private|emergency_access"]'
                                      '["motor_vehicle"!~"no"]["motorcar"!~"no"]'
                                      )
                        )
fig, ax = ox.plot_graph(G, edge_linewidth=0.5, node_size=0, save=True,filename='shang', file_format='svg')
```
